```python
import pandas as pd
import numpy as np

nba = pd.read_csv("basketball_players.csv")
print(nba.head(3))

players = pd.read_csv("basketball_master.csv")
```

```python
# Limited datasets
tmp = nba[nba['year'] > 2010]
nba = tmp[tmp["lgID"] == "NBA"]
print(nba.head(1))

#Index into the players to convert from ID to name with:
#print(players[players['bioID'] == "alexaco02"])

print("Size of dataset: ", len(nba))
print(nba.columns)
```

```python
# Add in a PPG column
nba['ppg'] = nba['points'] / nba['GP']

# Eliminate any 0 turnover players, and then calculate Assists - Turnover ratio
nba = nba[nba['turnovers'] != 0]
nba['atr'] = nba['assists'] / nba['turnovers']

# Let's set a reasonable ~10 minutes per game minimum
nba = nba[nba['minutes'] > 1600]

nba = nba[['points', 'GP', 'ppg', 'assists', 'turnovers', 'atr']]
```

```python
#Sanity check- let's make sure the columns are right
nba.head(2)
print(len(nba))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import re
plt.style.use('ggplot')
plt.rcParams.update({'font.size': 16})

# Visualize
fig = plt.figure(num=None, figsize=(6, 6), edgecolor='k')

ax = fig.add_subplot(1, 1, 1, facecolor="1.0")
ax.scatter(nba['ppg'], nba['atr'], alpha=0.5, c='y', edgecolors='g', s=150)
plt.title("NBA players in 2011")
plt.xlabel('Points Per Game')
plt.ylabel('Assist Turnover Ratio')
plt.show()
```

```python
# K-means clustering by hand

num_clusters = 5

# Select 5 random players initially
random_initial_points = np.random.choice(nba.index, size=num_clusters)
centroids = nba.loc[random_initial_points]

fig = plt.figure(num=None, figsize=(6, 6), edgecolor='k')

ax = fig.add_subplot(1, 1, 1, facecolor="1.0")

ax.scatter(nba['ppg'], nba['atr'], alpha=0.5, c='y', edgecolors='g', s=150)
ax.scatter(centroids['ppg'], centroids['atr'], marker='x')

plt.title("Initial Centroids")
plt.xlabel("PPG")
plt.ylabel("ATR")
plt.show()
```

```python
# Let's make it easier to track the centroids

def centroids_to_dict(centroids):
    dictionary = {}
    
    counter = 0
    
    for index, row in centroids.iterrows():
        coordinates = [row['ppg'], row['atr']]
        dictionary[counter] = coordinates
        counter += 1
        
    return dictionary

# Now we have a dictionary of numbered clusters
centroids_dict = centroids_to_dict(centroids)
```

```python
# A quick Euclidean distance function for determining the cluster
import math

def calculate_distance(centroid, player_values, normalized=False):
    root_distance = 0
    
# The biggest issue is that the clustering all ends up being vertical
# columns as the difference in PPG is so large compared to the ATR
# so we normalize the values first
    ppg_std = nba['ppg'].std()
    atr_std = nba['atr'].std()
    
# Sloppy but it works
    normalized = (ppg_std, atr_std)
    
    for x in range(0, len(centroid)):
        difference = (centroid[x] - player_values[x]) / normalized[x]
        squared_difference = difference**2
        root_distance += squared_difference
        
    euclid_distance = math.sqrt(root_distance)
    return euclid_distance
```

```python
# Iterate through each cluster and assign to the closest one
def assign_to_cluster(row):
    player_vals = [row['ppg'], row['atr']]
    dist_prev = -1
    cluster_id = None
    
    for centroid_id, centroid_vals in centroids_dict.items():
        dist = calculate_distance(centroid_vals, player_vals, normalized=True)
        if dist_prev == -1:
            cluster_id = centroid_id
            dist_prev = dist
        elif dist_prev > dist:
            cluster_id = centroid_id
            dist_prev = dist
            
        #print("Distance to cluster ", centroid_id, " from ", player_vals, " to ", centroid_vals, " is ", dist)
    
    return cluster_id

# Now let's apply it to every row
nba['cluster'] = nba.apply(lambda row: assign_to_cluster(row), axis=1)
print(nba.head(2))
            
```

```python
# Visualize these clusters

def visualize_clusters(df, num_clusters, iteration):
    colors = ['b', 'y', 'r', 'c', 'y', 'm']
    fig = plt.figure(num=None, figsize=(6, 6), edgecolor='k')
    ax = fig.add_subplot(1, 1, 1, facecolor = "1.0")
    for n in range(num_clusters):
        clustered_df = df[df['cluster'] == n]
        ax.scatter(clustered_df['ppg'], clustered_df['atr'], c=colors[n-1], edgecolors='g', alpha=.5, s=150)
    
    for centroid in centroids_dict.values():
        ax.scatter(centroid[0], centroid[1], marker='x', c='black')

    plt.xlabel('Points Per Game')
    plt.ylabel('Assist Turnover Ratio')
    plt.title('Iteration Number %s'%(iteration)) 
    
    plt.show()
    
iteration = 0
visualize_clusters(nba, 5, iteration)
```

```python
# Now we want the new center of the centroids
def recalculate_centroids(df):
    new_centroids_dict = {}
    
    for cluster_id in range(num_clusters):
        df_cluster_id = df[df['cluster'] == cluster_id]
        
        xmean = df_cluster_id['ppg'].mean()
        ymean = df_cluster_id['atr'].mean()
        
        new_centroids_dict[cluster_id] = [xmean, ymean]
        
    return new_centroids_dict
    
centroids_dict = recalculate_centroids(nba)    
```

```python
# Iterate 3 times
nba['cluster'] = nba.apply(lambda row: assign_to_cluster(row), axis=1)
visualize_clusters(nba, num_clusters, 1)

centroids_dict = recalculate_centroids(nba)
nba['cluster'] = nba.apply(lambda row: assign_to_cluster(row), axis=1)
visualize_clusters(nba, num_clusters, 2)

centroids_dict = recalculate_centroids(nba)
nba['cluster'] = nba.apply(lambda row: assign_to_cluster(row), axis=1)
visualize_clusters(nba, num_clusters, 3)
```

```python
# Let's standardize data for k-means- or as above, we will get a ton of clustering along the ATR axis as
# that is so small compared to PPG
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

nba['ppg_scaled'] = min_max_scaler.fit_transform(nba[['ppg']])
nba['atr_scaled'] = min_max_scaler.fit_transform(nba[['atr']])

print(nba.head(2))
```

```python
# Now lets do it with sklearn
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=num_clusters, max_iter=100)

kmeans.fit(nba[['ppg_scaled', 'atr_scaled']])

nba['cluster'] = kmeans.labels_

visualize_clusters(nba, num_clusters, 100)
```
